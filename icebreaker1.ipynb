```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords# Import the library
import pandas as pd
```

```python
resume_skills=pd.read_csv('resume_data.csv')
jobs=pd.read_csv('Top30.csv')
```

```python
resume_skills.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer()
skill = input('Enter skils:')#Summarize the text with ratio 0.1 (10% of the total words.)
summarize(skill, ratio=0.1)
per=dict()
pers=[]
for i in range(0,70000):
    jd=jobs['Description'][i]
    jd = str(jd)#Summarize the text with ratio 0.1 (10% of the total words.)
    text_list = [skill, jd]    
    count_matrix = cv.fit_transform(text_list)
    matchPercentage = cosine_similarity(count_matrix)[0][1] * 100
    matchPercentage = round(matchPercentage, 2) # round to two decimal
    per[i+1]=matchPercentage#value:percentage. We need sorting based on the percentage values
sorted_values = sorted(per.values()) # Sort the values
sorted_dict = dict()
for j in sorted_values:
    for k in per.keys():
        if per[k] == j:
            sorted_dict[k] = per[k]
            break
k=list(sorted_dict.keys())
print(jobs['Query'][k[-10:]])
#Keymax = max(per, key=per.get) 
#print(jobs['Query'][Keymax-1])
#print(Keymax)
```
